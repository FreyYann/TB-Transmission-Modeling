```python
# y'= -sin(t), y(0)= 1.
t0= 0    #start point
te= 10    #end point
delta= 0.1    #step length

y= np.zeros(101)    #declare funtion value
dy= np.zeros(101)    #declare derivative value
y[0]= 1;    #initial value

for i in np.arange(100):
    dy[i]= -np.sin(i*delta);    #computing the first derivative
    y[i+1]= y[i]+ delta*dy[i];    #estimating function value by differential property

time= np.arange(t0,te+delta,delta);    

plt.plot(time, y,'r') #plotting
    
sint= [np.cos(t) for t in time]
plt.plot(time,sint,'b')
plt.xlabel('time t')
plt.ylabel('cos')
plt.title('Plotting differential equations.')
plt.legend(('cos_definite','cos_standard'))
plt.ylim([-1.2,1.2])
plt.show()
```

$ \frac { dS }{ dt } =A-\lambda SI-dS\\ 
\frac { dE }{ dt } =(1-\rho )\lambda SI-\nu E-dE\\ 
\frac { dI }{ dt } =\rho \lambda SI+\nu E+\eta R-dI-\mu I-\gamma I\\ 
\frac { dR }{ dt } =\gamma I-\eta R-dR$

```python
x=range(16)
pop=np.array(df_china.e_pop_num.tolist())#whole population
X=np.array(df_china.e_inc_num.tolist())##new TB infected cases each year
S=pop-X#suspetable population
d=0.0067
A=1.623*pow(10,7)
v=6
m=0.0025
p=0.05
r=0.496
ti=0.51
fei=0.9
n=0.00341
lam=2.553*pow(10,-9)
```

```python
##simulation of x(t) where dx(t)/dt=p*S*I+v*E+n*R
## z(t) =x(t)-x(t-1) represents newly-infected TB cases
t0= 0    #start point
te= 16    #end point
delta=0.1    #step length

x= np.zeros(161)  
dx= np.zeros(161) 

I_=np.zeros(161)
dI_=np.zeros(161)

S_= np.zeros(161)
dS_= np.zeros(161)

E_=np.zeros(161)
dE_=np.zeros(161)

R_=np.zeros(161)
dR_=np.zeros(161)

x[0]= 0;  #initial value
I_[0]= X[0];
S_[0]=S[0]
E_[0]=117403
R_[0]=776223


for i in np.arange(160):
    
    dx[i]=p*(1-fei)*lam*S_[i]*I_[i]+v*E_[i]+n*R_[i];    #computing the first derivative
    
    dS_[i]=A-lam*S_[i]*I_[i]-d*S_[i]
    dE_[i]=(1-p)*(1-fei)*lam*S_[i]*I_[i]-v*E_[i]-d*E_[i]
    dI_[i]=p*lam*(1-fei)*S_[i]*I_[i]+v*E_[i]+n*R_[i]-d*I_[i]-m*I_[i]-(1+ti)*r*I_[i]
    dR_[i]=(1+ti)*r*I_[i]+n*R_[i]-d*R_[i]
    
    S_[i+1]=S_[i]+ delta*dS_[i]
    E_[i+1]=E_[i]+ delta*dE_[i]
    I_[i+1]=I_[i]+ delta*dI_[i]
    R_[i+1]=R_[i]+ delta*dR_[i]
    x[i+1]= x[i]+ delta*dx[i];    #estimating function value by differential property

time= np.arange(t0,te+delta,delta); 
```

```python
Z=[x[t*10]-x[t*10-1*10] for t in range(1,16)]
Time=range(16)
plt.plot(Time[1:], Z,'r') #plotting
y= [X[t] for t in Time]
plt.plot(Time,y,'b')
plt.xlabel('time t')
plt.ylabel('yearly new TB cases ')
plt.title('Plotting differential equations.')
plt.legend(('test','real'))
#plt.ylim([-1.2,1.2])
plt.show()
```
